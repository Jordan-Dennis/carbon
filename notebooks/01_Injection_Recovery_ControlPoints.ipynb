# Recovery test on simulated Noisy data 
In this notebook we simulated a noisy d14c data from a sine production rate function and try to recover the production rates using control points interpolated by a Gaussian Process. The control points are first fitted using the L-BFGS-B method from `scipy.minimize.optimize`, with objective function being the negative chi2 likelihood of prediction and the true d14c data + negative Gaussian Process likelihood. Then, the fitted control points are sampled with MCMC to explore the uncertainties in the fitted control points. 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams

import ticktack
import jax.numpy as jnp
from jax import jit
from ticktack import fitting
import emcee

rcParams['figure.figsize'] = (10.0, 5.0)
```

## Simulating Noisy d14c data
Start by defining a parametric production rate function, this will give the true production rates. For this notebook, we use a sine function with a 11-year period. In the next cell, the attributes for a `CarbonFitter` have been defined explicitly. In practice, one can just call `cf.load_data(filename.csv)`, provided the csv file is in the required structure (see example csv file).

```python
@jit
def sine(t):
    prod =  1.87 + 0.7 * 1.87 * jnp.sin(2 * jnp.pi / 11 * t + jnp.pi/2)
    prod = prod * (t>=cf.start) + (1.87 + 0.18 * 1.87 * jnp.sin(2 * jnp.pi / 11 * cf.start + jnp.pi/2)) * (1-(t>=cf.start))
    return prod
```

```python
cbm = ticktack.load_presaved_model('Guttler14', production_rate_units = 'atoms/cm^2/s')
cf = fitting.CarbonFitter(cbm)
cf.prepare_function(custom_function=True, f=sine)
cf.time_data = jnp.arange(200, 230) 
cf.d14c_data_error = jnp.ones((cf.time_data.size,))
cf.start = np.nanmin(cf.time_data)
cf.end = np.nanmax(cf.time_data)
cf.resolution = 1000
cf.burn_in_time = jnp.linspace(cf.start-1000, cf.start, cf.resolution)
cf.time_grid_fine = jnp.arange(cf.start, cf.end, 0.05)
cf.time_oversample = 1000
cf.offset = 0
cf.gp = True
cf.annual = jnp.arange(cf.start, cf.end + 1)
cf.mask = jnp.in1d(cf.annual, cf.time_data)[:-1]
```

Visualizing the true production rates,

```python
plt.plot(cf.time_data, sine(cf.time_data), ".b", markersize=7)
plt.plot(cf.time_data, sine(cf.time_data), "b")
plt.ylabel("Production rate ($cm^2s^{-1}$)");
plt.xlabel("Calendar Year");
plt.title("Sine production rate");
```

Visualizing true d14c data and noisy d14c data. Adding unit gaussian noise on the d14c data gives SNR ~10.

```python
np.random.seed(0)
d14c = cf.dc14()
noisy_d14c = np.array(d14c) + np.random.randn(d14c.size) # add unit gaussian noise
noisy_d14c = np.append(noisy_d14c, noisy_d14c[-1]) # for compatibility with ticktack code
plt.errorbar(cf.time_data[:-1], d14c, yerr=cf.d14c_data_error[:-1], fmt="-k", fillstyle="none", capsize=2, label="d14c")
plt.plot(cf.time_data[:-1], noisy_d14c[:-1], ".r", label="noisy d14c", markersize=10)
plt.ylabel("$\Delta^{14}$C (‰)")
plt.xlabel("Calendar Year");
plt.legend();
```

Here we have the noisy d14c data stored in `cf.d14c_data`. Based on this noisy data, we want to recover the true production rate as best as we can. A non-parametric approach is using control points, which determine the shape of the production function. To avoid long runtime, we use only one control point per 2 years. 

```python
cf.d14c_data = jnp.array(noisy_d14c)
cf.prepare_function(use_control_points=True, interp='gp')
cf.control_points_time = jnp.arange(cf.start, cf.end, 2)
print("%d years in the period, %d control-points used" % (cf.annual.size, cf.control_points_time.size))
```

Fitting the control points,

```python
%%time
soln = cf.fit_cp(low_bound=0.)
```

```python
plt.plot(cf.control_points_time, soln.x, ".b", markersize=7, label="recovered production rate")
plt.plot(cf.time_grid_fine, sine(cf.time_grid_fine), 'k', label='true production rate')
plt.title("Recovery test");
plt.ylabel("Production rate ($cm^2s^{-1}$)");
plt.xlabel("Calendar Year");
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.13), fancybox=True)
```

```python
plt.errorbar(cf.time_data[:-1], d14c, yerr=cf.d14c_data_error[:-1], fmt="-k", fillstyle="none", capsize=2, label="true d14c")
plt.plot(cf.time_data[:-1], noisy_d14c[:-1], ".r", label="noisy d14c", markersize=10)
plt.plot(cf.time_data[:-1], cf.dc14(soln.x), ".b", label="recovered d14c", markersize=10)
plt.title("Recovery test");
plt.ylabel("$\Delta^{14}$C (‰)");
plt.xlabel("Calendar Year");
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.13), fancybox=True)
```

Sampling fitted control points with MCMC sampler from `emcee`

```python
sampler = cf.sampling(soln.x, cf.gp_sampling_likelihood, burnin=500, production=1000)
```

```python
cf.plot_recovery(sampler, time_data=cf.time_grid_fine, true_production=sine(cf.time_grid_fine))
```
