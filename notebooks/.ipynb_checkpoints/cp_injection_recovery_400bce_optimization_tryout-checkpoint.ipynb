```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import rcParams
import warnings
warnings.filterwarnings('ignore')

import re
import jax.numpy as jnp
from jax import grad, jit, partial
import ticktack
from ticktack import fitting
from astropy.table import Table
from tqdm import tqdm
from jax.ops import index_update
import scipy

rcParams['figure.figsize'] = (10.0, 5.0)
```

```python
cbm = ticktack.load_presaved_model('Guttler14', production_rate_units = 'atoms/cm^2/s')
cf = fitting.CarbonFitter(cbm)
cf.load_data('400_BCE_Data_processed.csv')
cf.prepare_function(use_control_points=True)
t = np.arange(np.nanmin(cf.time_data)-1, np.nanmax(cf.time_data)+1)
```

```python
steady_state = cf.steady_state_production * jnp.ones((cf.end-cf.start+2,))
offset = jnp.array([0.])
params = jnp.append(steady_state, offset)
bounds = tuple([(0, None)] * len(steady_state) + [(0, None)])
```

```python
def f(optimiser):
    soln = scipy.optimize.minimize(cf.loss_chi2_fit_offset, params, method=optimiser, bounds=bounds, tol=1e-8, options={"disp":True})
    print(cf.loss_chi2_fit_offset(params=soln.x))
    fig, (ax1, ax2)  =plt.subplots(2, figsize=(10, 10))
    ax1.plot(t, soln.x[:-1])
    ax1.set_ylabel("Production rate ($cm^2s^{-1}$)");
    ax1.set_xlabel("Calendar Year");
    ax1.set_title(r"Fitted control-points with avg");
    
    fine = cf.dc14_fine_FitOffset(soln.x)
    coarse = cf.dc14_FitOffset(soln.x)
    ax2.plot(cf.time_data[:-1], coarse, "ob", fillstyle="none", label="fitted")
    ax2.plot(cf.time_grid_fine[:-1], fine, ".", label="fine")
    ax2.errorbar(cf.time_data, cf.d14c_data, yerr=cf.d14c_data_error, fmt=".k", label="true")
    ax2.legend()
    ax2.set_ylabel("$\Delta^{14}$C (â€°)")
    ax2.set_xlabel("Calendar Year");
    ax2.set_title("$\Delta$14C data");
```

```python
f("Nelder-Mead")
```
